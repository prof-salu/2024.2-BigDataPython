```python
#Convertendo tipos de dados no dataframe

#Year_of_Release --> int
#Critic_Count --> int
#User_Count --> int

df = df.astype({
    'Year_of_Release' : 'int',
    'Critic_Count' : 'int',
    'User_Count' : 'int'
})
```

```python
df.head()
```

```python
#Ajustando a escala da coluna User_Score
df['User_Score'] = df['User_Score'] * 10
```

```python
#Remover as informações sobre vendas:
#NA_Sales	EU_Sales	JP_Sales	Other_Sales	Global_Sales
#Remover as informações sobre Critic e User Count

df.drop(['NA_Sales', 'EU_Sales', 'JP_Sales',
         'Global_Sales','Other_Sales', 'Critic_Count',
         'User_Count'], axis=1, inplace=True)
```

```python
#Todos os jogos do PS4, mas apenas as colunas:
# Name, Year e Developer

df.loc[df['Platform'] == 'PS4',
 ['Name', 'Year_of_Release', 'Developer']]
```

```python
#Todos os jogos do PS4 com nota maior que 80,
#mas apenas as colunas: Name, Year e Developer
df.loc[(df['Platform'] == 'PS4')
& (df['Critic_Score'] > 80),
 ['Name', 'Year_of_Release', 'Developer']]
```

```python
#Os 10 melhores jogos do PS3,
#mas apenas as colunas: Name, Year e Developer
df.loc[df['Platform'] == 'PS3',
       ['Name',
        'Year_of_Release',
        'Developer',
        'Critic_Score']].sort_values(
            ['Critic_Score'],
            ascending=[False]).head(10)
```

```python
#Gerando uma nova coluna para o dataframe
#Media das notas Critic e User
df['AVG_Score'] = (df['Critic_Score'] + df['User_Score']) / 2
```
