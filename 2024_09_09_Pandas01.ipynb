#Pandas

```python
#Biblioteca python para trabalhar com um conjunto de dados.

#importar a biblioteca do pandas
import pandas as pd
```

```python
#Existem dois objetos principais no Pandas
# - DataFrames [similar tabelas]
# - Series [similar a uma coluna de uma tabela]
```

```python
#Criando um dataframe
cores_de_carros = pd.DataFrame({
    'Uno' : ['Verde', 'Vermelho', 'Cinza'],
    'Gol' : ['Azul', 'Preto', ''],
    'Onix' : ['Branco', 'Preto', 'Cinza']
}, index=['Filial RJ', 'Filial SP', 'Filial BH'])
```

```python
print(cores_de_carros)
```

```python
#Criar um data de frame sobre votos

votos = pd.DataFrame({
    'sim': [110, 21, 34],
    'nao': [44, 77, 99],
    'talvez': ['Não informado', 55, 100]
}, index=['Grupo 1', 'Grupo 2', 'Grupo 3'])

display(votos)
```

```python
#Series

vendas = pd.Series([101, 33, 77, 89],
                   index=['2015', '2016', '2017', '2018'],
                   name='Produto A')

display(vendas)
```
